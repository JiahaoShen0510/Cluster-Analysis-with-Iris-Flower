```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.cluster import KMeans
from sklearn import preprocessing
```

```python
# import and organize the dataset
```

```python
data0=pd.read_csv("iris_dataset.csv")
data0
```

```python
data0.isnull().count()
```

```python
# cluster analysis without standardization
```

```python
x=data0[["sepal_length","sepal_width","petal_length","petal_width"]]
```

```python
kmeans=KMeans()
```

```python
kmeans.fit(x)
```

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
wcss=[]

for i in range (1,151):
    kmeans=KMeans(i)
    kmeans.fit(x)
    wcss_inet=kmeans.inertia_
    wcss.append(wcss_inet)
```

```python
wcss
```

```python
number_clusters=range (1,151)
plt.plot(number_clusters,wcss)
plt.xlabel("Clusters",fontsize=15)
plt.ylabel("Sum of Squares of the Distances",fontsize=15)
plt.xlim(0,10)
```

```python
kmeans01=KMeans(2)
```

```python
kmeans01.fit(x)
```

```python
plt.scatter(data0["sepal_length"],data0["sepal_width"],c=kmeans01.fit_predict(x),cmap="rainbow")
plt.xlabel("Sepal Length:2 clusters",fontsize=15)
plt.ylabel("Sepal Width: 2clusters",fontsize=15)
```

```python
plt.scatter(data0["petal_length"],data0["petal_width"],c=kmeans01.fit_predict(x),cmap="rainbow")
plt.xlabel("Petal Length:2 clusters",fontsize=15)
plt.ylabel("Petal Width:2 clusters",fontsize=15)
```

```python
kmeans02=KMeans(3)
```

```python
kmeans02.fit(x)
```

```python
plt.scatter(data0["sepal_length"],data0["sepal_width"],c=kmeans02.fit_predict(x),cmap="rainbow")
plt.xlabel("Sepal Length:3 clusters",fontsize=15)
plt.ylabel("Sepal Width: 3 clusters",fontsize=15)
```

```python
plt.scatter(data0["petal_length"],data0["petal_width"],c=kmeans02.fit_predict(x),cmap="rainbow")
plt.xlabel("Petal Length:3 clusters",fontsize=15)
plt.ylabel("Petal Width: 3 clusters",fontsize=15)
```

```python
# cluster analysis with the standardized data
```

```python
x_scaled=preprocessing.scale(x)
```

```python
kmeans_scaled=KMeans()
```

```python
kmeans_scaled.fit(x_scaled)
```

```python
wcss_scaled=[]

for i in range (1,151):
    kmeans_scaled=KMeans(i)
    kmeans_scaled.fit(x_scaled)
    wcss_inet_scaled=kmeans_scaled.inertia_
    wcss_scaled.append(wcss_inet_scaled)
```

```python
wcss_scaled
```

```python
plt.plot(number_clusters, wcss_scaled)
plt.xlabel("Clusters",fontsize=15)
plt.ylabel("Sum of Squares of the Distances",fontsize=15)
plt.xlim(0,20)
```
